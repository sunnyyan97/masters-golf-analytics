```python
import duckdb
import pandas as pd

conn = duckdb.connect('data/masters.duckdb')

# See all tables
conn.execute("SHOW TABLES").df()
```

```python
conn.execute("""
    SELECT player_name, win_pct, top5_pct, top10_pct, mc_pct
    FROM mart_simulation_results
    ORDER BY win_pct DESC
    where model_type = 'regression'
    LIMIT 20
""").df()
```

```python
conn.execute("""WITH reg AS (
    SELECT player_name, win_pct, top10_pct, mu,
           row_number() OVER (ORDER BY win_pct DESC) AS reg_rank
    FROM mart_simulation_results
    WHERE model_type = 'regression'
),
man AS (
    SELECT datagolf_id, win_pct, top10_pct, mu,
           row_number() OVER (ORDER BY win_pct DESC) AS manual_rank
    FROM mart_simulation_results
    WHERE model_type = 'manual'
),
inputs AS (
    SELECT datagolf_id, player_name,
           sg_arg, sg_app, sg_ott, sg_putt,
           driving_dist_vs_avg, long_approach_sg,
           augusta_sg_total, augusta_seasons_played,
           owgr_rank
    FROM mart_player_model_inputs
)
SELECT
    reg.reg_rank                             AS reg_rank,
    man.manual_rank                          AS manual_rank,
    (man.manual_rank - reg.reg_rank)         AS rank_delta,   -- positive = regression ranks higher
    i.player_name,
    round(reg.win_pct  * 100, 1)             AS reg_win_pct,
    round(man.win_pct  * 100, 1)             AS manual_win_pct,
    round(i.sg_arg,  3)                      AS sg_arg,
    round(i.sg_app,  3)                      AS sg_app,
    round(i.sg_ott,  3)                      AS sg_ott,
    round(i.driving_dist_vs_avg, 1)          AS dist_vs_avg,
    round(i.long_approach_sg, 3)             AS long_app_sg,
    round(i.augusta_sg_total, 3)             AS aug_career_sg,
    i.augusta_seasons_played                 AS aug_seasons,
    i.owgr_rank
FROM reg
JOIN man    USING (player_name)
JOIN inputs i USING (player_name)
ORDER BY reg.reg_rank
LIMIT 30;
""").df()
```
