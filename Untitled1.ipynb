```python
import duckdb
import pandas as pd

conn = duckdb.connect('data/masters.duckdb')

# See all tables
conn.execute("SHOW TABLES").df()
```

```python
conn.execute("""
    SELECT player_name, augusta_fit_score,
           PERCENT_RANK() OVER (ORDER BY augusta_fit_score DESC) as pct_rank
    FROM mart_player_model_inputs
    ORDER BY augusta_fit_score DESC
    LIMIT 20
""").df()
```

```python
conn.execute("""
    SELECT player_name, win_pct, top5_pct, top10_pct, mc_pct
    FROM mart_simulation_results
    where model_type = 'regression'
    ORDER BY win_pct DESC
    
    LIMIT 50
""").df()
```

```python
conn.execute("""
    SELECT player_name, win_pct, top5_pct, top10_pct, mc_pct
    FROM mart_simulation_results
    where model_type = 'manual'
    ORDER BY win_pct DESC
    LIMIT 20
""").df()
```

```python
conn.execute("""
    SELECT 
        model_type,
        COUNT(*) as years,
        ROUND(AVG(spearman_corr), 3) as avg_spearman,
        ROUND(AVG(top10_precision), 3) as avg_top10_pct,
        ROUND(MIN(spearman_corr), 3) as worst_year,
        ROUND(MAX(spearman_corr), 3) as best_year
    FROM mart_backtest_comparison
    GROUP BY model_type
    ORDER BY avg_spearman DESC
""").df()
```

```python
conn.execute("""
    SELECT 
        year,
        model_type,
        ROUND(spearman_corr, 3) as spearman,
        ROUND(top10_precision, 3) as top10_pct,
        winner_rank,
        notes
    FROM mart_backtest_comparison
    ORDER BY year, spearman_corr DESC
""").df()
```

```python
import duckdb
import pandas as pd
from scipy.stats import spearmanr

conn = duckdb.connect('data/masters.duckdb')

# Get both model predictions
df = conn.execute("""
    SELECT 
        b.year,
        b.model_type,
        b.spearman_corr,
        b.top10_precision
    FROM mart_backtest_comparison b
    WHERE b.model_type IN ('manual', 'regression')
    ORDER BY b.year, b.model_type
""").df()

# Pivot and compute ensemble as simple average
pivot = df.pivot(index='year', columns='model_type', values='spearman_corr')
pivot['ensemble_avg'] = (pivot['manual'] + pivot['regression']) / 2

print(pivot)
print(f"\nEnsemble avg Spearman: {pivot['ensemble_avg'].mean():.3f}")
print(f"Manual avg Spearman:   {pivot['manual'].mean():.3f}")
print(f"Regression avg Spearman: {pivot['regression'].mean():.3f}")
```
